```python
import spacy

nlp = spacy.load('en_core_web_sm')
text = "hello, #welcome123 @to NLP"

def clean_text(text):
    cleaned_text = "".join(char for char in text if char.isalpha() or char.isspace())
    doc = nlp(cleaned_text)
    return " ".join(token.text for token in doc)

clean_str = clean_text(text=text)
print(clean_str)
```

```python
chat_word_dict = {
    "u": "you",
    "r": "are",
    "b4": "before",
    "gr8": "great",
    "l8r": "later",
    "idk": "I don’t know",
    "btw": "by the way",
    "imo": "in my opinion",
    "ttyl": "talk to you later",
    "brb": "be right back",
    "wanna": "want to",
    "gonna": "going to",
    "lemme": "let me",
    "dunno": "don’t know",
    "thx": "thanks",
    "pls": "please",
    "ppl": "people",
    "smh": "shaking my head",
    "tbh": "to be honest",
    "omg": "oh my god",
    "lol": "laugh out loud",
    "rofl": "rolling on the floor laughing",
    "afk": "away from keyboard",
    "np": "no problem",
    "gtg": "got to go",
    "jk": "just kidding",
    "hbu": "how about you",
    "ily": "I love you",
    "ttys": "talk to you soon"
}
```

```python
chat_text = "omg idk what un talking about lol"



def normalize_text(chat_text):
    words = chat_text.split()
    normalized_words = [chat_word_dict.get(word.lower(), word) for word in words]
    return " ".join(normalized_words)


```

```python
normalized_text=normalize_text(chat_text = "omg idk what ur talking about lol")
```
